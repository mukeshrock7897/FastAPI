## âœ… Using `TestClient` (FastAPI Built-in)

Use `TestClient` to test your FastAPI app **without starting a real server**.

```python
from fastapi import FastAPI
from fastapi.testclient import TestClient

app = FastAPI()

@app.get("/ping")
def ping():
    return {"message": "pong"}

# Create a test client
client = TestClient(app)

# Test function
def test_ping():
    response = client.get("/ping")
    assert response.status_code == 200
    assert response.json() == {"message": "pong"}
```

âœ… Best for: **Quick route testing** during development.

## ğŸ­ Mocking with `dependency_overrides`

Use `app.dependency_overrides` to **swap real logic** (like DB/auth) with **fakes** in tests.

### ğŸ§© Real Dependency

```python
# main.py
from fastapi import Depends

def get_db():
    return {"source": "prod-db"}

@app.get("/data")
def fetch_data(db = Depends(get_db)):
    return db
```

### ğŸ§ª Mock It in Test

```python
# tests/test_main.py
from main import app, get_db
from fastapi.testclient import TestClient

client = TestClient(app)

# Fake DB for testing
def fake_db():
    return {"source": "test-db"}

# Override real with fake
app.dependency_overrides[get_db] = fake_db

def test_fetch_data():
    res = client.get("/data")
    assert res.status_code == 200
    assert res.json() == {"source": "test-db"}
```

âœ… Useful for **unit tests**, **faster tests**, and **no real DB calls**.
