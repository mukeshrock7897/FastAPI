# üí¨ 9. **WebSocket Support in FastAPI**

FastAPI makes it **super simple** to build real-time apps using WebSockets for features like **live chat**, **real-time dashboards**, **collaboration tools**, and **IoT monitoring**.

---

## üì° 9.1 WebSocket Endpoint

Define a route using `@app.websocket()` to handle real-time connections.

```python
from fastapi import FastAPI, WebSocket

app = FastAPI()

@app.websocket("/ws/chat")
async def websocket_endpoint(websocket: WebSocket):
    await websocket.accept()
    await websocket.send_text("üëã Welcome to the chat!")
    while True:
        data = await websocket.receive_text()
        await websocket.send_text(f"Echo: {data}")
```

üìå Connect to `ws://localhost:8000/ws/chat` using a WebSocket client.

## üßæ 9.2 `websocket.accept()`

Accept the incoming WebSocket connection before interacting:

```python
await websocket.accept()
```

üîê Useful to perform auth checks *before* accepting.

## üîÑ 9.3 `send_text()` & `receive_text()`

### Sending messages:

```python
await websocket.send_text("üì¢ Hello Client!")
```

### Receiving messages:

```python
data = await websocket.receive_text()
print(f"Received: {data}")
```

## üß† Example Use Case: Real-Time Chat

```python
@app.websocket("/ws")
async def chat_room(websocket: WebSocket):
    await websocket.accept()
    while True:
        msg = await websocket.receive_text()
        await websocket.send_text(f"You said: {msg}")
```
