# üí¨ WebSocket in FastAPI

Use WebSockets to enable **real-time features** like live chat, notifications, dashboards, or multiplayer games ‚Äî all without page reloads.

---

## üì° Basic WebSocket Example

```python
from fastapi import FastAPI, WebSocket

app = FastAPI()

@app.websocket("/ws/chat")
async def websocket_endpoint(websocket: WebSocket):
    await websocket.accept()
    await websocket.send_text("üì¢ Connected to server!")
    
    while True:
        msg = await websocket.receive_text()
        await websocket.send_text(f"üîÅ You said: {msg}")
```

‚úÖ Test this using:

* **Browser Console**
* **VS Code WebSocket plugin**
* **CLI tool**: `websocat ws://localhost:8000/ws/chat`

## üí¨ Real Chat Handler

```python
@app.websocket("/ws")
async def chat(websocket: WebSocket):
    await websocket.accept()
    while True:
        text = await websocket.receive_text()
        await websocket.send_text(f"‚úÖ Received: {text}")
```

### üñ•Ô∏è Client-side JavaScript

```js
const socket = new WebSocket("ws://localhost:8000/ws");

socket.onmessage = (e) => console.log("üì©", e.data);

socket.onopen = () => {
    socket.send("Hello FastAPI!");
};
```

‚úÖ This runs inside the browser and connects to your FastAPI server.

## üîê Optional: Auth Before Accept

Use query params or headers for simple authentication before accepting the connection.

```python
@app.websocket("/ws")
async def secure_ws(websocket: WebSocket):
    token = websocket.query_params.get("token")
    if token != "secret123":
        await websocket.close()
        return

    await websocket.accept()
```

> ‚ùå Closes the connection if the token is invalid.
