## ğŸ”„ 8.1 Middleware (Custom Logic)

Use middleware to run code **before and after every request**.

```python
from fastapi import FastAPI, Request

app = FastAPI()

@app.middleware("http")
async def log_requests(request: Request, call_next):
    print(f"ğŸ“¥ Incoming: {request.method} {request.url}")
    response = await call_next(request)
    print(f"ğŸ“¤ Outgoing: {response.status_code}")
    return response
```

ğŸ“Œ Great for logging, metrics, or modifying requests/responses.

## ğŸŒ 8.2 CORS Middleware (Cross-Origin Resource Sharing)

Allow **frontend apps** (e.g., React, Vue) on different domains to access your FastAPI backend.

```python
from fastapi.middleware.cors import CORSMiddleware

app.add_middleware(
    CORSMiddleware,
    allow_origins=["*"],  # or ["http://localhost:3000"]
    allow_credentials=True,
    allow_methods=["*"],
    allow_headers=["*"],
)
```

ğŸ“Œ Must-have for public APIs or full-stack apps.

Perform actions **when the app starts or stops**, like DB setup or resource cleanup.

```python
@app.on_event("startup")
async def startup():
    print("âœ… App is starting...")
    # connect to DB or load models

@app.on_event("shutdown")
async def shutdown():
    print("ğŸ›‘ App is shutting down...")
    # close DB or save logs
```

ğŸ“Œ Perfect for ML model loading, DB pooling, or cleaning up temp files.
