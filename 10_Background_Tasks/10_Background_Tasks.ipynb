## âš™ï¸ 10.1 `BackgroundTasks` Class

ğŸ“Œ Import and inject via FastAPI to run background functions **after response is sent**.

```python
from fastapi import FastAPI, BackgroundTasks

app = FastAPI()

def write_log(message: str):
    with open("log.txt", "a") as f:
        f.write(message + "\n")

@app.post("/submit/")
async def submit_data(data: str, background_tasks: BackgroundTasks):
    background_tasks.add_task(write_log, f"Data received: {data}")
    return {"message": "Submitted!"}
```
