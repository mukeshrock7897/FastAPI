## ğŸ“¦ 3.1 What is `BaseModel`?

Use `BaseModel` to **define data structure** for input/output (like forms or JSON).

```python
from pydantic import BaseModel
from fastapi import FastAPI

app = FastAPI()

class User(BaseModel):
    id: int
    name: str
    email: str

@app.post("/user")
def create_user(user: User):
    return user
```

ğŸ’¡ This:

* Checks types (`int`, `str`, etc.)
* Returns clean JSON
* Shows model in API docs

## ğŸ§° 3.2 Validating Fields with `Field()`

You can **add rules** like minimum length, required field, greater than, etc.

```python
from pydantic import BaseModel, Field

class Product(BaseModel):
    name: str = Field(..., min_length=2, max_length=50)
    price: float = Field(..., gt=0)
    in_stock: bool = True
```

ğŸ’¡ Notes:

* `...` = required
* `min_length`, `max_length` = text length rules
* `gt`, `lt` = greater than / less than for numbers

## ğŸ” 3.3 Nested Models (Models Inside Models)

Used when your data has sub-objects (like address inside user).

```python
class Address(BaseModel):
    city: str
    zip: str

class Customer(BaseModel):
    name: str
    address: Address
```

ğŸ“Œ Example input:

```json
{
  "name": "Khushi",
  "address": {
    "city": "Bangalore",
    "zip": "560001"
  }
}
```

âœ… FastAPI handles this automatically.

## ğŸ§¾ 3.4 `Enum` and `Union` Types

### âœ… Enum â†’ Fixed options only

```python
from enum import Enum

class Size(str, Enum):
    small = "S"
    medium = "M"
    large = "L"

class Shirt(BaseModel):
    size: Size
```

ğŸ¯ Only allows `"S"`, `"M"`, `"L"` â€” prevents invalid input.

### ğŸ”„ Union â†’ Allow different types

```python
from typing import Union

class InputData(BaseModel):
    data: Union[int, str]
```

âœ… Input can be a number **or** text.

## ğŸ›¡ï¸ 3.5 Custom Validation with `@validator`

Use this when default rules arenâ€™t enough. Example: no spaces in username.

```python
from pydantic import validator

class User(BaseModel):
    username: str

    @validator("username")
    def no_spaces_allowed(cls, value):
        if " " in value:
            raise ValueError("No spaces allowed")
        return value
```

ğŸ’¡ Useful for:

* Email or password rules
* Format checking
* Business logic
