## âœ… 7.1 Routers (`APIRouter`)

ğŸ”¹ Use `APIRouter` to **split endpoints** into modular sections.

```python
# routers/items.py
from fastapi import APIRouter

router = APIRouter()

@router.get("/items")
def get_items():
    return ["Item 1", "Item 2"]
```

ğŸ”¹ Register the router in your main app:

```python
# main.py
from fastapi import FastAPI
from routers import items

app = FastAPI()
app.include_router(items.router, prefix="/api")
```

ğŸ“Œ Makes it easy to organize endpoints like `/api/items`, `/api/users`, etc.
